```python
def plot_distribution_of_distances(this_G):
    con_dict = networkx.all_pairs_node_connectivity(this_G)
    all_distances = []
    for node_id, node_connections_dict in con_dict.items():
        for distance in node_connections_dict.values():
            all_distances.append(distance)
    plt.hist(all_distances);
```

```python
plot_distribution_of_distances(G3)
```
